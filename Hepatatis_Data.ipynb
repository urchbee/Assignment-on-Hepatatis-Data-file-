```python
import pandas as pd
```

```python
#import hepatatis csv file by uploading on github.
#copying raw data from github repository, call it by assigning it to variable url
url = 'https://raw.githubusercontent.com/urchbee/Assignment-on-Hepatatis-Data-file-/main/hepatitis-1.csv'

#to bring out the file to work on it, define dataframe and read csv(url)
df1 = pd.read_csv(url)
df1
```

```python
#to check the general information of the data such as the data type  
df1.info()
```

```python
#to check for any missing values in the data as this is an important step in cleaning the data. 
missing_values = df1.isnull().sum()
missing_values
#in the output, it shows there are some colums in the data that have missing values
```

```python
#cleaning the data further by dropping columns that have missing values
#we do this by using dropna funtion and calling axis=1

hepatatis = df1.dropna(axis=1)
hepatatis
#this prints out the data without the columns that have any missing values
```

```python
#to ensure the data is clean of any missing values, we use isnull once again
missing_values1 = hepatatis.isnull().sum()
missing_values1

```

The data has now been clean by removing all missing values hence the data is now uniform. an advantage of cleaning this data is that it has improved the quality of the data
